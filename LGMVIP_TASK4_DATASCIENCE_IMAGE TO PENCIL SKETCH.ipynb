```python
import cv
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('cat.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,6))
plt.imshow(img)
plt.axis('off')
plt.title('original image')
```

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,6))
plt.imshow(gray_image,cmap='gray')
plt.axis('off')
plt.title('Gray image')
plt.show()
```

```python
image_invert = cv2.bitwise_not(gray_image)
plt.figure(figsize=(8,6))
plt.imshow(image_invert,cmap='gray')
plt.axis('off')
plt.title('inverted Image')
plt.show()
```

```python
Gsn_image = cv2.GaussianBlur(image_invert,(21,21),sigmaX=0,sigmaY=0)
plt.figure(figsize=(8,6))
plt.imshow(Gsn_image,cmap='gray')
plt.axis('off')
plt.title('Gaussian Blur Image')
plt.show()
```

```python
pencil_sketch = cv2.divide(gray_image,255 - Gsn_image, scale = 255)
plt.figure(figsize=(10,8))
plt.imshow(pencil_sketch,cmap='gray')
plt.axis('off')
plt.title('Pencil_Sketch Image')
plt.show()
```
